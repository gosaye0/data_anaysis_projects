## Data exploration of raw datasets

For the exploration part we need pandas package

```python
import pandas as pd
```

```python
# Load all datasets for exploration
customers = pd.read_csv("../data/raw/olist_customers_dataset.csv")
orders = pd.read_csv("../data/raw/olist_orders_dataset.csv")
products = pd.read_csv("../data/raw/olist_products_dataset.csv")
product_category = pd.read_csv("../data/raw/product_category_name_translation.csv")
sellers = pd.read_csv("../data/raw/olist_sellers_dataset.csv")
order_items = pd.read_csv("../data/raw/olist_order_items_dataset.csv")
order_payments = pd.read_csv("../data/raw/olist_order_payments_dataset.csv")
order_reviews = pd.read_csv("../data/raw/olist_order_reviews_dataset.csv")
geolocation = pd.read_csv("../data/raw/olist_geolocation_dataset.csv")
```

```python
# Initial data exploration
print("Shape of each datasets")
print(f"Customers: {customers.shape}")
print(f"orders: {orders.shape}")
print(f"products: {products.shape}")
print(f"product_category: {product_category.shape}")
print(f"sellers: {sellers.shape}")
print(f"order_items: {order_items.shape}")
print(f"order_payments: {order_payments.shape}")
print(f"order_reviews: {order_reviews.shape}")
print(f"geolocation: {geolocation.shape}")
```
